## What is Activation Function? 
* When you are building your neural network, one of the design decisions that you will need to 
make is what activation function to use for our neurons calculations. 


* Activation function is also 
referred to as **transfer function** or **nonlinearities** because they transform the linear combination 
of the weighted sum into non-linear models. It is placed at the end of each perceptron to 
decide whether to activate this neuron or not. 

**Why use activation functions at all? Why not just calculate the weighted sum of our 
network and propagate that through the hidden layers to produce an output?**

> The purpose of the activation function is to introduce non-linearity into the network. Without 
it, a multi-layer perceptron will perform similar to a single perceptron no matter how many 
layers we add. Activation functions are needed to restrict the output value to a certain finite 
value.

## Activation Functions

* Activation Function apply a Non_Linear_Transfomation and decide whether a neuron should be active or dead. 
* Without Non Linearity in our NN, our nodes are like a simple stacked linear regression. 
* A simple linear model doesn't learn anything, and it can't solve more problems. 


#### Main Thing! 

* It add non linearity to nodes. 

### 1. Sigmoid Activation Function: 

* Sigmoid Usually used in Binary Classification Problems. (0,1) 
* It should be in last layer of your neural network. 
* The sigmoid squishes all the 
values to a probability between 0 and 1 which reduces extreme values or outliers in the data 
without removing them. 
* Sigmoid or logistic functions converts infinite continuous variables 
(range between ‚àí‚àû to +‚àû) into simple probabilities between 0 and 1. 
* It is also called the ‚ÄúS‚Äù
shape curve because when plotted in a graph it produces an S shaped curve.
* In **forward propagation** the values should be **0** and **1**. 
* In **backward propagation** the values should be **0** to **0.25**.

### 2. Threshold Activation Function ( Tan h): 

* It is a shifted version of the sigmoid version. Instead of squeezing the signals values between 
0 and 1, tanh squishes all values to the range -1 to 1. 
* Tanh almost always works better than 
the sigmoid function in the hidden layers because it has the effect of centring your data so 
that the mean of the data is close to zero rather than 0.5 which makes learning for the next 
layer a little bit easier. 
* Tan h ranges between **-1 to +1** during forward propagation. 
* Tan h ranges between **0 to 1** during backward propagation. (derivative)
* It is **Zero Centric**. 
* It is a good choice for hidden layers. 
* It is better than sigmoid.

### Problem: 
* One of the downsides for both sigmoid and tanh functions is that, if (z) is very large or very 
small, then the gradient (or derivative or slope) of this function becomes very small (close to 
zero) which will slow down gradient descent.
* Vanishing Gradient Problem.
* It take more time for computation. 

### 3. Rectifier Linear Unit (ReLU): 

* In Both **Sigmoid and Tan h** activation function ( after back propagation ), the actual value and derivative value will be same ( vanishing gradient descent )after some number of iterations. To overcome this, we use ReLU. 
* ReLU activation function activates a node only if the input is above zero. If the input is below 
zero, the output is always zero. 

#### Formula:

**RELU = MAX(0, X)**


* RELU : Whenver your value is greater than 0, you will get the respective value, if your value is lesser than 0 then the output will be lesser than 0. 

* Derivative: If your value is more than 0(x),the derivative is dx/dx =1, if your value is lesser than 0, the derivative will be 0. 

* It is widely used in **hidden layers** becuase it solves the **vanishing gradient problems** HOw? because the derivative is either 0 or 1. 

* The values are either 0(negative) or infinity( x value ) 


### Problem: 

* Whenever you have a negative value during back propagation value is negative value right!. So, if it calculate the derivative, it becomes zero so it causes the dead neuron. 
* It is called dying relu or dead activation function.

### 4. Leaky Relu (LReLU): 
* To Overcome the Dead Neuron or died activation function, we will be use Leaky relu activation function. 

* It prevent the dead neuron. How? same formula but here we will add 0.01 to the higher value of x, so it just overcome the ReLU dead neuron problem. 

* Instead of having 
the function being zero when `x < 0`, Leaky ReLU introduces a small negative slope (around 
0.01) when (x) is negative.

* In simple words, the derivative will not be 0 it will add .1 to your derivative so your neurons will be alive. 

* The value are  -infinit(0.01) to + infinity(positive value) 

**Why 0.01? Some people like to use this as another hyperparameter to tune, but that would be 
an overkill since you already have other bigger problems to worry about. Feel free to try 
different values (0.1, 0.01, 0.002) in your model and see how they work.**

### 9. Soft Max: 
* Softmax function is a generalization of the sigmoid function. It is used to obtain classification 
probabilities when we have more than 2 classes. It is used to force the outputs of Neural 
Network to sum to 1. 
* The most common use case in deep learning 
(especially in computer vision) is to predict a single class out of many options (more than 2). 
* Whenever we have 2 output layers, we use sigmoid activation function. 
* If we have more than 2 output, we can use **SOFT MAX** activation function. 
* It is always in the last layer!
* Softmax function is the go to function that you will often use at the output layer of a classifier when you 
are working on a problem that you need to predict a class between more than 2 classes. Softmax can work 
fine if you are classifying two classes as well. It will basically work as a sigmoid function.
